```python
def new_column(df, original, new):
    df[new]=df[original]
    return df
```

```python
#discnew=new_column(discipline, 'ACTION_TAKEN', 'recommended_discipline')
#discnewer=new_column(discnew, )
discipline['recommended_discipline']= discipline['CODE']
discipline['final_discipline']= discipline['CODE']
discipline['recommended_action']=discipline['ACTION_TAKEN']
discipline['final_action']=discipline['ACTION_TAKEN']
discipline
```

```python
def disc_merge(disc,accused,original,new):
    disc_new=new_column(disc, original, new)
    acc_merge=pd.merge(accused, disc_new, new)
    return acc_merge
```

```python
accrec=pd.merge(complaints_accused,discipline[['recommended_discipline','recommended_action']], 
                how='left',on='recommended_discipline', suffixes=('_acc','_disc'))

accrec
```
